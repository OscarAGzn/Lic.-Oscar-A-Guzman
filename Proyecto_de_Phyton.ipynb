#La Tuberculosis (TB)
###1. Definición y Agente Causal
La tuberculosis es una enfermedad infecciosa y potencialmente grave causada por la bacteria Mycobacterium tuberculosis. Aunque afecta predominantemente a los pulmones (tuberculosis pulmonar), el patógeno tiene la capacidad de diseminarse a través del torrente sanguíneo o el sistema linfático a otros órganos, como los riñones, la columna vertebral y el cerebro (tuberculosis extrapulmonar).

###2. Mecanismo de Transmisión
La vía de contagio es estrictamente aérea. Cuando una persona con la enfermedad activa en los pulmones o la laringe tose, estornuda, habla o canta, expulsa al aire pequeñas gotas (núcleos de Wells) que contienen los bacilos. Para que ocurra el contagio, generalmente se requiere un contacto cercano y prolongado en espacios mal ventilados. Es importante notar que la TB no se transmite por compartir cubiertos, dar la mano o el contacto con superficies.

###3. Infección Latente vs. Enfermedad Activa
Es fundamental distinguir entre dos estados:

Infección de TB Latente: Las bacterias viven en el cuerpo pero el sistema inmunitario las mantiene bajo control. La persona no presenta síntomas, no se siente enferma y no puede contagiar a otros.

Enfermedad de TB Activa: Las bacterias se multiplican porque el sistema inmune no logra contenerlas. La persona presenta síntomas clínicos y puede transmitir la bacteria a los demás.

###4. Cuadro Clínico y Diagnóstico
Los síntomas clásicos de la TB pulmonar incluyen:

Tos persistente (más de 3 semanas), a menudo acompañada de esputo o sangre.

Dolor torácico.

Fatiga crónica y debilidad.

Pérdida de peso involuntaria y anorexia.

Fiebre y sudores nocturnos profusos.

El diagnóstico se confirma mediante pruebas de laboratorio como la baciloscopia (examen de flema), cultivos microbiológicos, pruebas moleculares rápidas (como GeneXpert) y radiografías de tórax para identificar lesiones pulmonares.

###5. Tratamiento y Resistencia a los Fármacos
La tuberculosis es una enfermedad curable. El tratamiento estándar dura entre 6 y 9 meses y requiere la combinación de varios antibióticos (isoniazida, rifampicina, etambutol y pirazinamida).

El mayor desafío actual para la salud pública es la TB Multirresistente (MDR-TB), la cual surge principalmente cuando los pacientes no completan el esquema terapéutico. Esto permite que las bacterias sobrevivan y desarrollen resistencia a los fármacos más potentes, obligando a tratamientos mucho más largos, costosos y con mayores efectos secundarios.

###6. Prevención y Control
La estrategia principal de prevención incluye la vacunación con BCG en recién nacidos (que protege contra formas graves como la meningitis tuberculosa) y, sobre todo, la detección temprana y el tratamiento oportuno de los casos activos para cortar la cadena de transmisión en la comunidad.

```python
#Cono puedo convertir columnas en filas utilizando pandas? = existe la funcion melt()
#Buscar la funcion melt en la documentacion oficial
# Tengo una tabla de esta estructura
who2=Who.melt(id_vars=["country", "iso2","iso3","year"])
who2.head(10)
```

```python
# Crar una columna que contenga el genero (masculinon o femenino)
#Crear una fncion de Phyton que contenga las reglas
def asignar_genero(valor):
  if"m" in valor:
    return "Masculino"
  else:
    return "Femenino"
```

```python
who2['genero'] = who2['variable'].apply(asignar_genero)
who2.head(10)
```

```python
# the method of diagnosis (rel = relapse, sn = negative pulmonary smear, sp = positive pulmonary smear, ep = extrapulmonary)
def asignar_metodo(valor):
  if "rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapulmonary"
  else:
    return "to define"
```

```python
who2["metodo"] = who2["variable"].apply(asignar_metodo)
who2.head(10)
```

```python
# age group (014 = 0-14 yrs of age, 1524 = 15-24, 2534 = 25-34, 3544 = 35-44 years of age, 4554 = 45-54, 5564 = 55-64, 65 = 65 years or older).
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoeddad"] = who2["variable"].apply(asignar_grupoedad)
who2.head(10)
```

```python
who2.drop(columns=["variable"], inplace=True)
```

```python
who2.rename(columns={"value":"cases"}, inplace=True)
```
